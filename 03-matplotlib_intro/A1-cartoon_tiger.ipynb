```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Ellipse, Arc, Rectangle
def Ellipse_with_outline(center, width, height, color, angle=0, clip=None):
    ellipse = Ellipse(center, width, height, color=color, angle=angle, transform=ax.transData)
    outline = Ellipse(center, width, height, color='k', fill=False, lw=3, angle=angle)
    if clip:
        ellipse.set_clip_path(clip)
        outline.set_clip_path(clip)
    return [ellipse, outline]
def Mirror_Ellipse_with_outline(center, width, height, color, angle=0, clip=None):
    left = Ellipse_with_outline(center, width, height, color, angle, clip)
    right = Ellipse_with_outline((100-center[0], center[1]), width, height, color, -angle, clip)
    return left+right
ax = plt.gca()
mouth_box, mouth_box_outline = Ellipse_with_outline((50, 26), 40, 26, 'b')
mouth, mouth_outline = Ellipse_with_outline((50, 40), 25, 35, 'r', clip=mouth_box)
toungue = Ellipse_with_outline((50, 25), 20, 10, '#ee838b', clip=mouth)
mouth_box_outline.set_clip_path(mouth)
face, face_outline = Ellipse_with_outline((50, 50), 60, 60, '#fee002')
marks = [
    Ellipse((50,80), 4, 25, color='k'),
    Ellipse((20,55), 7,3, color='k'),
    Ellipse((20,45), 7,3, color='k', angle=20),
    Ellipse((80,55), 7,3, color='k'),
    Ellipse((80,45), 7,3, color='k', angle=-20)
]
for mark in marks:
    mark.set_clip_path(face)
shapes = [
    *Ellipse_with_outline((70, 70), 30, 24, color="#fee002", angle=45),    # ear
    *Ellipse_with_outline((70, 70), 15, 15, color="w", angle=45),    # ear inner    
    *Ellipse_with_outline((30, 70), 30, 24, color="#fee002", angle=-45),   # ear
    *Ellipse_with_outline((30, 70), 15, 15, color="w", angle=-45),    # ear inner
    face,     # face
    Ellipse((50, 25), 50, 40, color='w'),     # face white
    face_outline,    # face outline
    Ellipse((35, 50), 7, 13, color='k'),    # eye
    Arc((33, 63), 10, 10, theta1=45, theta2=180, color='k', lw=3),    # eyebrow
    Ellipse((65, 50), 7, 13, color='k'),   #  eye
    Arc((67, 63), 10, 10, theta1=0, theta2=135, color='k', lw=3),    # eyebrow
    Ellipse((34, 51), 2, 3.5, color='w'),  # eye ball
    Ellipse((64, 51), 2, 3.5, color='w'), # eye ball
    #mouth_box,
    mouth,
    *toungue,
    mouth_outline,
    mouth_box_outline,
    Ellipse((50, 43), 5, 3, color='#cd535b'),    # nose
    Rectangle((49, 39), 1, 3, color='k'),    # nose line
    Ellipse((50, 43), 5, 3, color='k', fill=False, lw=3),    # nose outline
    Ellipse((50,77), 18, 3, color='k'),
    Ellipse((50,72), 10, 4, color='k'),
    Arc((22, 20), 40, 40, theta1=80, theta2=105, color='k', lw=2),    # hair
    Arc((78, 20), 40, 40, theta1=75, theta2=100, color='k', lw=2),    # hair
    Arc((27, 18), 40, 40, theta1=95, theta2=115, color='k', lw=2),    # hair
    Arc((73, 18), 40, 40, theta1=65, theta2=85, color='k', lw=2),    # hair
    *marks
]
for shape in shapes:
    ax.add_patch(shape)
ax.set_xlim((10, 90))
ax.set_ylim((10, 90))
ax.set_aspect('equal', adjustable='datalim')
```
